In Python, the **property decorator** is used to define **"getter"** and **"setter"** methods for class attributes, allowing you to control access to and modification of those attributes. It is a way to implement computed or managed properties in a more controlled and Pythonic manner.

```python
class pwskills:
  def __init__(self, course_name, course_price):

    self.course_name = course_name
    self.__course_price = course_price
```

```python
pw = pwskills("data science masters", 3500)
```

```python
pw.course_price
```

```python
pw.course_name
```

```python
class pwskills1:
  def __init__(self, course_name, course_price):

    self.course_name = course_name
    self.__course_price = course_price

# give access to see the course price
  @property
  def course_price_access(self):
      return self.__course_price

```

```python
pw = pwskills1("data science masters", 3500)
```

```python
pw.course_price_access
```

```python
class pwskills2:
  def __init__(self, course_name, course_price):

    self.course_name = course_name
    self.__course_price = course_price

  @property
  def course_price_access(self):
      return self.__course_price

# give access to modify the course price

  @course_price_access.setter
  def course_price_set(self, price):
    if price <= 3500:
      pass
    else:
      self.__course_price = price

  @course_price_access.deleter
  def delete_course_price(self) :
      del self.__course_price
```

```python
pw = pwskills2("data science masters", 3500)
```

```python
pw.course_price_set = 4500
```

```python
pw.course_price_access
```

```python
pw.pwskills2__course_price
```

```python
del pw.delete_course_price
```
